```python
from langchain_core.prompts import ChatPromptTemplate

# Define the blueprint with placeholders
template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role} mentor."),
    ("human", "What is the best way to learn {topic}?")
])

# Fill the placeholders
formatted_prompt = template.invoke({
    "role": "Python",
    "topic": "programming"
})

# Print the formatted messages
print(formatted_prompt.to_messages())
```

```python
base_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {expert_type} mentor."),
    ("human","{question}")])

python_expert_prompt = base_template.partial(expert_type="Python")

final_prompt = python_expert_prompt.invoke({
    "question": "What is the best way to learn programming?"
})

print(final_prompt.to_messages())
```

```python
from langchain_core.prompts import ChatPromptTemplate

# Create the template
template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert translator from {input_language} to English."),
    ("human", "Translate this text:  {text}")
])

# Invoke the template
formatted_prompt = template.invoke({
    "input_language": "Marathi",
    "text": "तू कसा आहेस?"
})

print(formatted_prompt.to_messages())
```

```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

load_dotenv()

llm = ChatGoogleGenerativeAI(model= "gemini-2.5-flash", temperature=0.7, max_tokens=100)

chain = template | llm

result = chain.invoke({
    "input_language": "Marathi",
    "text": "तू कसा आहेस?"
})

print(result.content)
```
